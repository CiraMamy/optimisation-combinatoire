# Cas 1 : Optimisation d’un plan de production

## Objectif
Maximiser le bénéfice d’une entreprise produisant des poupées et des voitures.

## Méthode
- Modélisation mathématique
- Résolution avec PuLP

```python
import pulp

print("Bibliothèques chargées")
```

```python
# Création du problème
model = pulp.LpProblem("Maximisation_Profit", pulp.LpMaximize)
```

```python
# Variables
x1 = pulp.LpVariable("Poupées", lowBound=0)
x2 = pulp.LpVariable("Voitures", lowBound=0)
```

```python
# Fonction objectif
model += 4 * x1 + 6 * x2
```

```python
# Contraintes
model += 2 * x1 + 3 * x2 <= 3600  # plastique
model += 3 * x1 + 4 * x2 <= 2000  # machine
```

```python
model.solve()
```

```python
print("Statut :", pulp.LpStatus[model.status])

print("Poupées :", x1.varValue)
print("Voitures :", x2.varValue)
print("Profit maximum :", pulp.value(model.objective))
```

## Interprétation des résultats

Le modèle indique combien de poupées et de voitures produire pour maximiser le profit tout en respectant les contraintes de ressources.

Ces résultats permettent d’aider à la prise de décision dans un contexte industriel.
