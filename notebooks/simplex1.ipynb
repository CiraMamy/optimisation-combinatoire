# Cas 1 : Optimisation d’un plan de production

## Objectif
Maximiser le bénéfice d’une entreprise produisant des poupées et des voitures.

## Méthode
- Modélisation mathématique
- Résolution avec PuLP

```python
import pulp

print("Bibliothèques chargées")
```

```python
# Création du problème
model = pulp.LpProblem("Maximisation_Profit", pulp.LpMaximize)
```

```python
# Variables
x1 = pulp.LpVariable("Poupées", lowBound=0)
x2 = pulp.LpVariable("Voitures", lowBound=0)
```

```python
# Fonction objectif
model += 4 * x1 + 6 * x2
```

```python
# Contraintes
model += 2 * x1 + 3 * x2 <= 3600  # plastique
model += 3 * x1 + 4 * x2 <= 2000  # machine
```

```python
model.solve()
```

```python
print("Statut :", pulp.LpStatus[model.status])

print("Poupées :", x1.varValue)
print("Voitures :", x2.varValue)
print("Profit maximum :", pulp.value(model.objective))
```

```python
# Vérification des contraintes
print("\nVérification :")
print("Plastique utilisé =", 2*x1.varValue + 3*x2.varValue)
print("Machine utilisée =", 3*x1.varValue + 4*x2.varValue)
```

## Interprétation des résultats

Le modèle indique combien de poupées et de voitures produire pour maximiser le profit tout en respectant les contraintes de ressources.

Ces résultats permettent d’aider à la prise de décision dans un contexte industriel.

### 1. Présentation du problème

Dans cette étude, nous considérons une entreprise fabriquant deux types de produits : des poupées et des voitures. L’objectif est de déterminer les quantités à produire afin de maximiser le bénéfice total, tout en respectant les contraintes de ressources disponibles, notamment le plastique et le temps machine.

Ce problème est modélisé comme un programme linéaire.

### 2. Modélisation du problème

#### Variables de décision :
- x1 : nombre de poupées produites
- x2 : nombre de voitures produites

#### Fonction objectif :
Maximiser :
Z = 4x1 + 6x2

#### Contraintes :
- 2x1 + 3x2 ≤ 3600 (plastique)
- 3x1 + 4x2 ≤ 2000 (machine)
- x1 ≥ 0, x2 ≥ 0

### 4. Résultats

La solution optimale obtenue est :

- x1 = 0 (poupées)
- x2 = 500 (voitures)
- Profit maximal = 3000

La contrainte liée au temps machine est saturée, tandis que celle du plastique ne l’est pas.

### 5. Résolution avec Python (PuLP)

Le même problème a été résolu à l’aide du langage Python et de la bibliothèque PuLP.

Les résultats obtenus sont identiques à ceux obtenus avec Excel :

- x1 = 0
- x2 = 500
- Profit = 3000

### 6. Comparaison des méthodes

Les résultats obtenus avec Excel et Python sont identiques, ce qui valide la modélisation du problème.

Cependant :

- Excel est plus simple à utiliser pour des petits problèmes
- Python est plus flexible et adapté à des problèmes complexes ou automatisés

### 7. Interprétation

La solution indique qu’il est optimal de produire uniquement des voitures.

Cela s’explique par le fait que les voitures génèrent un profit plus élevé et utilisent de manière plus efficace la ressource limitante (temps machine).

Ainsi, l’entreprise doit concentrer sa production sur les voitures pour maximiser son bénéfice.

## Analyse des résultats

La solution optimale indique qu’il est préférable de produire uniquement des voitures (x2 = 500) et aucune poupée (x1 = 0).

La contrainte machine est saturée, ce qui signifie qu’elle constitue le facteur limitant de la production.

En revanche, la contrainte plastique n’est pas saturée, ce qui indique qu’il reste des ressources inutilisées.

Ainsi, la stratégie optimale consiste à exploiter entièrement la ressource critique (machine) en produisant le produit le plus rentable.

```python
import numpy as np
import matplotlib.pyplot as plt

# Variables
x = np.linspace(0, 1000, 100)

# Contraintes
plastique = (3600 - 2*x) / 3
machine = (2000 - 3*x) / 4

# Graphique
plt.figure()

plt.plot(x, plastique, label="2x1 + 3x2 ≤ 3600")
plt.plot(x, machine, label="3x1 + 4x2 ≤ 2000")

# Zone faisable
plt.fill_between(x, 0, np.minimum(plastique, machine), alpha=0.3)

# Solution optimale
plt.scatter(0, 500)
plt.text(0, 500, "Optimum (0,500)")

plt.xlabel("Poupées (x1)")
plt.ylabel("Voitures (x2)")
plt.legend()
plt.title("Zone Faisable et Solution Optimale")

plt.show()
```
